## Bank Account System

```python
from pydantic import BaseModel
```

```python
account_info = {"owner": "John",
    "account_number": "123456",
    "balance": "5000.75",
    "is_active": "True",
    "transaction_history": ["500", "1000", "250"],
    "contact": {"email": "john@bank.com", "phone": "9999"}}
```

```python
class BankAccount(BaseModel):
    owner: str
    account_number: int
    balance: float
    is_active: bool
    transaction_history: list[float]
    contact: dict[str, str]
```

```python
def display_account(bankaccount: BankAccount):
    print(f"Owner: {bankaccount.owner}")
    print(bankaccount.account_number)
    print(bankaccount.balance)
    if bankaccount.is_active:
        print(f"Account Status: Active")
    else:
        print("Account Status: Inactive")
```

```python
bacc1 = BankAccount(**account_info)
```

```python
display_account(bacc1)
```
